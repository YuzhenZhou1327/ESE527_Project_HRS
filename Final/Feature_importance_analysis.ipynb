```python
import pandas as pd
import numpy as np
from scipy.stats import ttest_rel, pearsonr
from itertools import combinations
```

```python
df = pd.read_csv("output_feature_importances_combained.csv").drop(columns=["targets"])
groups = df["group"].unique()
df
```

```python
for i in combinations(groups, 2):
    a = df[df["group"] == i[0]].drop(columns=["group"]).values[0]
    b = df[df["group"] == i[1]].drop(columns=["group"]).values[0]
    t_test = ttest_rel(a, b)
    print(f"Comparing {i[0]} and {i[1]}")
    print(f"t-statistic: {t_test.statistic}, p-value: {t_test.pvalue}\n")
```

```python
correlated = {}

for i in combinations(df.drop(["group"], axis = 1).columns,2 ):
    a = df[i[0]].values
    b = df[i[1]].values
    correlation, p_value = pearsonr(a, b)
    if p_value < 0.05:
        print("!!!!!")
        correlated[str(i)] = correlation
    print(f"Comparing {i}")
    print(f"correlation: {correlation}, p-value: {p_value}\n")
```
