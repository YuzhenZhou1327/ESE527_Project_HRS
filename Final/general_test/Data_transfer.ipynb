```python
import pandas as pd
import numpy as np
from scipy.stats import boxcox
from sklearn.preprocessing import PowerTransformer
```

```python
df = pd.read_csv('test_data.csv')
df = df[df["HATOTB"] >= 0]
df_transformed = df.drop(['group'], axis=1)
keep_data = df[['group']]
```

```python
df_transformed.describe()
```

```python
df_log = df_transformed.apply(lambda x: np.log(x + 0.1))
df_log = pd.concat([df_log, keep_data], axis=1)
df_log
```

```python
df_log.describe()
```

```python
df_log.to_csv('test_dataset1_log.csv', index=False)
```
