```python
import pandas as pd
import numpy as np
from scipy.stats import boxcox
from sklearn.preprocessing import PowerTransformer
```

```python
df = pd.read_csv('ThreeForthSemester.csv')
keep_data = df[['SHLT','MSTOT','COGTOT','group']]
df_transformed = df.drop(['SHLT','MSTOT','COGTOT','group'], axis=1)
```

```python
min_shift = abs(df_transformed.min().min()) + 1
min_shift
```

```python
df_transformed.describe()
```

```python
df_log = df_transformed.drop(['HATOTB'],axis=1)
df_log = df_log.apply(lambda x: np.log(x + 0.1))
df_log = pd.concat([df_log, keep_data], axis=1)
df_log
```

```python
df_log.to_csv('ThreeForthSemester_log.csv', index=False)
```

```python
# Applying Box-Cox Transformation
data_box_cox = df_transformed.copy()
data_box_cox = data_box_cox.drop(['HATOTB'],axis = 1)
for col in data_box_cox.columns:
    data_box_cox[col], _ = boxcox(data_box_cox[col] + 0.01) 
data_box_cox = pd.concat([data_box_cox, keep_data], axis=1)
```

```python
data_box_cox.to_csv('ThreeForthSemester_box_cox.csv', index=False)
```

```python
pt = PowerTransformer(method='yeo-johnson')
data_yeo_johnson = pd.DataFrame(pt.fit_transform(df_transformed), columns=df_transformed.columns)
data_yeo_johnson = pd.concat([data_yeo_johnson, keep_data], axis=1)
data_yeo_johnson.to_csv('ThreeForthSemester_yeo_johnson.csv', index=False)
```
