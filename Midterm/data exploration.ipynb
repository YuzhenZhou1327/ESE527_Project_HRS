```python
df = pd.read_csv('cleaned_extracted_data.csv')
continuous_features = ['SHLT', 'BMI', 'MSTOT', 'COGTOT', 'INHPFN', 'HHHRES', 'HCHILD', 'LIVSIB', 'HAIRA', 'HATOTB', 'IEARN', 'HITOT', 'PRPCNT']

# Determine the size of the grid
n = len(continuous_features)
cols = 3  # Number of columns you want in your figure
rows = n // cols + (n % cols > 0)  # Calculate rows needed

# Create a figure with subplots
plt.figure(figsize=(20, rows * 4))  # Adjust the figure size as necessary

for i, feature in enumerate(continuous_features, start=1):
    plt.subplot(rows, cols, i)
    plt.hist(df[feature], bins=30, edgecolor='k')  # You can adjust the number of bins
    plt.yscale("log")  # Keeping the log scale for y-axis
    plt.title(feature)

plt.tight_layout()
plt.show()
```
